# Part 2 - Basic EDA

EDA stands for Exploratory Data Analysis

## 2.1 Using visuals in Python to explore the relationship between mosquito number and date.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

When exploring the relationship between two variables, one of which is a date and the other is a numeric variable such as mosquito number, a time series plot is often the most effective visualization. 

For this purpose, I can use a line plot or a scatter plot with the date variable on the x-axis and the mosquito number on the y-axis.

```python
# Plot a line plot
df.plot(x='Date', y='Mosquito number', kind='line', figsize=(16,8))
plt.title('Mosquito Number Over Time')
plt.xlabel('Date')
plt.ylabel('Mosquito Number')
plt.show()
```

The visualization above is not explaining well the relationships between the variables of Mosquito Number and Dates.

Going back to the Dictionary of Mosquito Data.
In the varible of Mosquito Numbe the test results are organized in such a way that when the number of mosquitoes exceed 50, they are split into another record (another row in the dataset), such that the number of mosquitoes are capped at 50.

If the goal is to visualize the relationship between the variable Date and the Mosquito Number (count), and I have multiple records for dates where the mosquito count exceeds 50, a step line plot may not effectively convey this relationship. 

In this case, I can try a different approach, such as grouping the data by Date and summing up the mosquito counts for each date, then plotting a regular line plot. 

This will give me a smoother visualization of the overall trend while still accounting for the splitting of records when the mosquito count exceeds 50.

```python
# Group the data by date and sum up the mosquito counts
daily_counts = df.groupby('Date')['Mosquito number'].sum()

# Plot a line plot
plt.figure(figsize=(16, 8))
plt.plot(daily_counts.index, daily_counts.values)
plt.title('Total Mosquito Number Over Time')
plt.xlabel('Date')
plt.ylabel('Total Mosquito Number')
plt.show()
```

This line plot shows the total mosquito count for each date (year), effectively visualizing the overall trend over time while accounting for the splitting of records when the mosquito count exceeds 50.

From 2016 to 2017 the count of mosquito number is higher.
While from 2008 to 2010, the count of mosquito number is lower.

It might be a relationship with another factors from those years that increased or decreased the mosquito number.

# Part 3 - Advanced EDA

## 3.1 Using visuals to explore the relationship between Mosquito Species and WNV Prevalence.

```python
# This code creates a new figure with a specified size of 10 inches in width and 6 inches in height using Matplotlib's plt.figure() function.
plt.figure(figsize=(10, 6))

# This code creates a count plot using Seaborn's sns.countplot() function. The ax variable stores the Axes object.
ax = sns.countplot(data=df, x='Species', hue='WNV Present')

# Add labels with counts above the bars
# The for loop is used to iterate over each bar in the plot and add labels with counts.
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Add title and labels
plt.title('WNV Prevalence by Mosquito Species')
plt.xlabel('Mosquito Species')
plt.ylabel('Count')

# Add legend
plt.legend(title='WNV Present', loc='upper right', labels=['No', 'Yes'])

# Show plot
plt.show()
```

This plot visualize the distribution of WNV prevalence among different mosquito species. 

The bars represent the count of occurrences for each species, and the colors indicate the presence or absence of WNV.

This visualization allows for easy comparison between species and helps identify any patterns or trends in WNV prevalence.

## 3.2 Using visuals to explore the relationship between Mosquito Species and WNV Prevalence.


To account for the fact that the number of mosquitoes is capped at 50 and may be split into multiple records, I will first group the data by 'Trap type' and then calculate the sum of the 'Mosquito number' for each trap type.

```python
# Group the data by 'Trap type' and sum the 'Mosquito number'
grouped_df = df.groupby('Trap type')['Mosquito number'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(16, 8))
sns.barplot(data=grouped_df, x='Trap type', y='Mosquito number')

# Add title and labels
plt.title('Total Mosquito Number by Trap Type')
plt.xlabel('Trap Type')
plt.ylabel('Total Mosquito Number')

# Show plot
plt.show()
```

This visualization allows you to compare the total number of mosquitoes caught by each trap type.

We have more mosquitos caught on the Gravid trap, the difference between trap types is drastic.
